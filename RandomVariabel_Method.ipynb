```python
import pandas as pd
class Method:
    
    

    # linear Cograential Generator[LGG]
    def LGG( a=6, m=100, c=37, x=25, n_random=6):
        Xn = []; Rn = []
        Xn.append(x); Rn.append(x/m)
        for i in range(n_random):
            x=(a*x+c)%m
            Xn.append(x)
            Rn.append(x/m)
        out = pd.DataFrame(data=Xn ,columns=['Xn'])
        out['Rn'] = Rn
        return out

    # multiplication Can  Generator[MGG]
    def MGG(x=25, a=6, m=127, n_random=6):
        Xn = []; Rn = []
        Xn.append(x); Rn.append(x/m)
        for i in range(n_random):
            x=(a*x)%m
            Xn.append(x)
            Rn.append(x/m)
        out = pd.DataFrame(data=Xn ,columns=['Xn'])
        out['Rn'] = Rn
        return out

    # mid square method
    def MSM(x=7612, m=10000, n_random=6):
        Xn = []; XnS = []; Rn = []; sq = x*x; div = x/m
        Xn.append(x); XnS.append(sq); Rn.append(div)
        if(len(str(x))>3):
            for i in range(n_random):
                size = len(str(x)); sq = str(sq); s = sq.split()
                size = len(str(sq)); size = int(size/2); 
                left = s[0][:size]; right =s[0][size:]
                orig = len(str(x)); div_orig = int(orig/2)
                num = left[div_orig:]+right[:div_orig]
                x = int(num); Xn.append(x); sq = x*x
                XnS.append(sq); div = x/m; Rn.append(div)
            out = pd.DataFrame(data=Xn ,columns=['Xn'])
            out['X_square'] = XnS
            out['Rn'] = Rn
            return out 
        else:
            print('something wrong!')
        
    
    
    
    
    
 
```
